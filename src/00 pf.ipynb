```python
# pip install pyarrow

import pandas as pd
import glob
import os
```

```python
# 1. Definir la ruta donde están tus archivos
# 1. Define the path where your files are located.
path = '../data/raw/' 

# 2. Listar todos los archivos que empiecen por 'data' y terminen en '.parquet'
# Esto busca data1.parquet, data2.parquet, etc.
# 2. List all files that begin with 'data' and end with '.parquet'
# This searches for data1.parquet, data2.parquet, etc.
all_files = glob.glob(os.path.join(path, "data*.parquet"))

# 3. Leer cada archivo y meterlo en una lista de DataFrames
# 3. Read each file and put it into a list of DataFrames
df_list = []
for filename in all_files:
    df_part = pd.read_parquet(filename, engine='pyarrow')
    df_list.append(df_part)
    print(f"Cargado: {filename} con {len(df_part)} filas")

# 4. Concatenar todos los DataFrames en uno solo
# 4. Concatenate all DataFrames into one
df = pd.concat(df_list, axis=0, ignore_index=True)

print(f"\nConcatenación completa. Tamaño total del dataset: {df.shape}")

# Mostrar las primeras filas para empezar el EDA
# Show the first few rows to start the EDA
df.head()
```

```python
#Selección de características. 
#ejemplo dado por kaggle
#Feature selection.
#Example provided by Kaggle

# Technical indicators
technical_features = [
    'SMA_5', 'SMA_10', 'RSI', 'MACD', 
    'BB_Position', 'Volatility'
]

# Lagged features
lag_features = [col for col in df.columns if 'lag' in col]

# All targets
targets = [col for col in df.columns if 'Future_' in col]
```

```python
# Ejemplo de entrenamiento del modelo. 
# #ejemplo dado por kaggle
# Model training example.
# #example provided by kaggle

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit

# Prepare features and target
features = technical_features + lag_features
X = df[features].fillna(method='ffill')
y = df['Future_Return_1d']

# Time series split
tscv = TimeSeriesSplit(n_splits=5)
model = RandomForestRegressor(n_estimators=100)

# Train model
for train_idx, test_idx in tscv.split(X):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
```
