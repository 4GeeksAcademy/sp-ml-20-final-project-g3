```python
zip_path = r"../data/processed/split_obj1.zip"
data_vars = {}

with zipfile.ZipFile(zip_path, 'r') as z:
    for nombre in z.namelist():
        if nombre.endswith('.csv'):
            with z.open(nombre) as f:
                # Extraemos el nombre sin el .csv
                llave = nombre.replace('.csv', '')
                data_vars[llave] = pd.read_csv(f)
                f"Cargada llave: {llave}"
```

```python
# Asignamos las variables usando los nombres exactos que están dentro del ZIP
# (quitando el .csv que hicimos en el bucle anterior)

X_train = data_vars.get('X_train_obj1')
X_test  = data_vars.get('X_test_obj1')
y_train = data_vars.get('y_train_obj1')
y_test  = data_vars.get('y_test_obj1')

# Verificación correcta (sin paréntesis y con los nombres nuevos)
if X_train is not None:
    f"X_train cargado correctamente. Forma: {X_train.shape}"
else:
    "Error: X_train sigue siendo None. Revisa si la llave 'X_train_obj1' existe en data_vars."
```
